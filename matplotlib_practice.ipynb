### plot chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[0,1,2,3,4]
y=[0,2,4,1,5]

plt.title("abc")
plt.plot(x,y,label="score")
plt.plot(x+x,y+y,label="score^2")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.legend()
plt.show()
```

### bar chart

```python
x=["a","b","c","d"]
y=[10,30,40,20]

plt.bar(x,y)

for i,v in enumerate(y):
  plt.text(i,v,str(v),ha="center")#각 위치별로 값 출력함
plt.ylim([0,50])
```

```python
plt.style.use('ggplot')

x=["a","b","c","d"]
y=[10,30,40,20]

plt.bar(x,y)

for i,v in enumerate(y):
  plt.text(i,v,str(v),ha="center")#각 위치별로 값 출력함
plt.ylim([0,50])
```

### 다중플롯, 서브플롯

```python
x=[0,1,2,3,4]
y=[0,1,4,9,16]
z=[0,-1,-4,-9,-16]
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,z)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,z)
```
