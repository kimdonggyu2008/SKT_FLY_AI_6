- iloc - 숫자로 위치 지정
- loc - 문자로 위치 지정

```python
df.iloc[0,0] #칼럼, 로우 기준으로 위치를 보고 지정
```

```python
df.loc[0,"NAME"] #칼럼 이름을 직접 명시하고, 위치를 지정함
```

```python
df.loc[0,["NAME","gender"]] #칼럼 이름을 직접 명시하고, 위치를 지정함, 여러개는 리스트로 입력
```

```python
df.loc[0:2,["NAME","gender"]] #칼럼 이름을 직접 명시하고, 위치를 지정함, 여러개는 슬라이스, 리스트로 입력
```

```python
df.loc[df['AGE']>26,["NAME","gender"]]
```

```python
df["AGE2"]=df['AGE']*2
```

```python
df
```

```python
df['AGE']=df["AGE"]//2
```

```python
df
```

```python
df.drop(["AGE2"],axis=1)
```

```python
df.drop(["AGE2"],axis=1,inplace=True)# inplace로 원본 교체도 가능
```

```python
df
```

```python
df = df.dropna(subset=["AGE"])#비어있는 값이 있으면 통째로 날려버림
```

```python
df
```

```python
df.fillna(0)#na에 몽땅 0으로 들어감, 타입 안 따짐
```

```python
df['AGE'].fillna(0)#특정 칼럼만 0으로 결측치 채워줌
```

### 데이터 집계 및 그룹핑


```python
data={
    'NAME':['ALICE','BOB','CHARLIE'],
    'AGE':[32,14,77],
    'gender':['male','female','male'],
    'else':["it","it","yet"]
}
df=pd.DataFrame(data)
df
```

```python
df.groupby('gender')['AGE'].sum()
```

```python
df.groupby('gender')['AGE'].mean()
```

```python
df['gender'].value_count()
```
