```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 라인플롯, 시퀸셜한 데이터에 따라 변하는 관계


```python
sns.set_theme(style='darkgrid')#테마 지정
```

```python
x=np.linspace(0,100,100)
```

```python
x
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
y
```

```python
z
```

```python
sns.lineplot(x=x[:50],y=y[:50])
sns.lineplot(x=x[:50],y=z[:50])
plt.title('data')
plt.show()#plt를 사용해서 sns의 설정을 지정할 수 있음
```

### 바 차트


```python
cat=['A','B','C','D']
val=[10,15,4,26]

sns.barplot(x=cat,y=val,hue=cat,palette='viridis')#hue 카테고리별로 따로따로 색상을 만듦
plt.ylim(0,30)
for i,v in enumerate(val):
  plt.text(i,v-2,str(v),fontsize=15, ha='center',color='white')
plt.show()
```

```python
cat=['A','B','C','D']
val=np.array([10,15,4,26])
val=np.sort(val)
sns.barplot(x=val,y=cat)
```

```python
data = sns.load_dataset('iris')

# 산점도 생성
sns.scatterplot(x=data['sepal_length'], y=data['sepal_width'],hue=data['species'])
```

```python
data.columns
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=data,hue='species',style='species')# 모든 데이터들을 다 사용해서 분류
#아래 그래프 상에서는 2차원 축이지만 정보가 3개(가로, 세로, 종류) 들어감
#hue를 넣으면 카테고리별 분류
#style을 넣으면 해당 카테고리별 다른 표시로 만들어줌
```

### 페어 플롯

```python
sns.pairplot(data=data,hue='species') #각 데이터셋에 대한 모든 관계를 전부 그려줌
```

### 박스플롯, 이상치 확인

```python
sns.boxplot(data=data,x=data['species'],y=data['sepal_width'],hue='species')#x는 카테고리, y는 연속형, 이산형 데이터 모두됨
```

### 바이올린 플롯

```python
sns.violinplot(data=data,x='species',y='sepal_width')
#불룩 튀어나오는 부분만큼 데이터 갯수가 많은 것임
```
