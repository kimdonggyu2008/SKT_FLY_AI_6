```python
#18년 4월부터 크롤링하면 될듯
import requests
import pandas as pd
import time
from datetime import datetime, timedelta

# ✅ API 기본 설정
api_url = "http://apis.data.go.kr/1230000/ad/BidPublicInfoService/getBidPblancListInfoServc"
#service_key = "xAI5Cagagsyktl+5DkdMACQ3R0c/seI5XjAcTth2misYLb5QZbxp0OZqYnCJ3/tGvlIeG6Ny0AjEszLbWitP3w=="
service_key = "X83Y2kuOIFofASmZXEDb6KTiVhheEug6mzsJVoTEoJZlLqX1zh3Ml8i+tQiuI4si0rN7aNjE8QB81AXbzN8V+w=="

# ✅ 저장할 CSV 파일명
output_csv_filename = "/content/drive/MyDrive/벽화_입찰_공고_2023_2024.csv"

# ✅ 2016년부터 2025년 1월까지 일주일 단위로 반복 조회
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 12, 31)
all_data = []  # 데이터를 저장할 리스트

while start_date <= end_date:
    # ✅ 조회 시작일 및 종료일 설정 (7일 단위)
    inqryBgnDt = start_date.strftime("%Y%m%d0000")
    inqryEndDt = start_date.strftime("%Y%m%d2359")
    start_date += timedelta(days=1)  # ✅ 1일 간격으로 증가

    print(f"\n🔍 조회 기간: {inqryBgnDt} ~ {inqryEndDt}")

    # ✅ API 요청 파라미터 설정
    params = {
        "serviceKey": service_key,
        "type": "json",
        "pageNo": "1",
        "numOfRows": "999",
        "inqryDiv": "1",  # 🔍 입찰공고 조회
        "inqryBgnDt": inqryBgnDt,
        "inqryEndDt": inqryEndDt,
    }

    try:
        # ✅ API 요청
        response = requests.get(api_url, params=params)
        print(f"🔗 요청 URL: {response.url}")

        if response.status_code == 200:
            data = response.json()

            # ✅ API 응답에서 items 가져오기
            items = data.get("response", {}).get("body", {}).get("items", [])

            if isinstance(items, list) and len(items) > 0:
                for item in items:
                    # ✅ "벽화" 키워드 포함된 공고만 필터링
                    if "벽화" in item.get("bidNtceNm", ""):
                        all_data.append(item)

            print(f"✅ {inqryBgnDt} ~ {inqryEndDt} 데이터 수집 완료 ({len(items)}건 중 '벽화' 포함: {len(all_data)}건)")

        else:
            print(f"❌ API 오류 발생: {response.status_code}")

        # ✅ API 요청 간격 조정 (0.5초 대기)
        time.sleep(2)

    except Exception as e:
        print(f"❌ 오류 발생: {e}")

# ✅ 데이터 저장
if all_data:
    df = pd.DataFrame(all_data)

    # ✅ 필요한 필드만 저장
    selected_columns = [
        "bidNtceNo", "bidNtceNm", "ntceInsttNm", "bidMethdNm",
        "cntrctCnclsMthdNm", "bidBeginDt", "bidClseDt", "opengDt",
        "presmptPrce", "bdgtAmt", "cnstrtsiteRgnNm"
    ]

    # ✅ 존재하는 컬럼만 선택하여 CSV 저장
    df = df[[col for col in selected_columns if col in df.columns]]
    df.to_csv(output_csv_filename, index=False, encoding="utf-8-sig")

    print(f"\n✅ 모든 '벽화' 관련 공고 저장 완료: {output_csv_filename}")
else:
    print("\n⚠️ '벽화' 관련 공고가 없습니다.")
```
