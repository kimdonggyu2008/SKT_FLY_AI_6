```python
img_gray=cv2.cvtColor(img_color,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img_gray)
```

```python
cv2.imwrite('gray_image.jpg',img_gray)
```

```python
h,w=img_color.shape[:2]
img_resized=cv2.resize(img_color,(w//2,h//2))
```

```python
cv2_imshow(img_resized)
```

```python
cv2.imwrite('resized_image.jpg',img_resized)
```

```python
img_hsv=cv2.cvtColor(img_color,cv2.COLOR_BGR2HSV)
```

```python
cv2_imshow(img_hsv)
```

```python
color_ranges={
    'red':{'lower':(0,100,100),'upper':(10,255,255)},
    'green':{'lower':(40,100,100),'upper':(70,255,255)},
    'blue':{'lower':(100,100,100),'upper':(130,255,255)}
}
```

```python
user_color='blue'
```

```python
img_hsv=cv2.cvtColor(img_color,cv2.COLOR_BGR2HSV)
```

```python
lower_val=color_ranges[user_color]['lower']
upper_val=color_ranges[user_color]['upper']
```

```python
mask=cv2.inRange(img_hsv,lower_val,upper_val)
```

```python
img_masked=cv2.bitwise_and(img_color,img_color,mask=mask)
```

```python
print(f" : {user_color}")
cv2_imshow(mask)
```

```python
cv2_imshow(img_masked)
```

```python
blurred_gaussian=cv2.GaussianBlur(img_color,(7,7),0)
cv2_imshow(blurred_gaussian)
cv2.imwrite('blurred_gaussian.jpg',blurred_gaussian)
```

```python
blurred_median=cv2.medianBlur(img_color,7)
cv2_imshow(blurred_median)
cv2.imwrite('blurred_median.jpg',blurred_median)
```
